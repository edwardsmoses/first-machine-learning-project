## Matplotlib Example Workflow

```python
# 0. Import Matplotlib and get it ready for plotting....

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# 1. Prepare some data

x = [1,2,3,4]
y = [11, 22, 33, 44]
```

```python
# 2. Setup Plots

fig, ax = plt.subplots(figsize=(10,10))
```

```python
#3. Plot the Data
ax.plot(x,y)
```

```python
#4. Customize the Plot

ax.set(title = "Simple Plot of X & Y Data", xlabel = "X Axis", ylabel = "Y Axis")
```

```python
fig
```

```python
# 5. Save & Show (Save the Figure to an Image File)
fig.savefig('images/sample-xy-plot.png')
```

```python
# Make a plot from Dictionary 

nut_butter_prices = {"Almond Butter" : 10, "Peanut Butter": 8, "Cashew Butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())

ax.set(title = "Dan's Nut Butter Store", ylabel = "Price ($)" );
```

```python
# Creating an horizontal Bar 
fig, ax = plt.subplots()

ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data

x = np.random.randn(1000)
```

```python
x[:10]
```

```python
# Create the Histogram using the Data...

fig, ax = plt.subplots()

ax.hist(x)
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
```

```python
car_sales
```

```python
# Remove last two zeros

car_sales["Price"] = car_sales["Price"].str[:-2]
```

```python
car_sales
```

```python
car_sales.Price, len(car_sales)
```

```python
car_sales["Sale Date"] = pd.date_range("7/7/2021", periods = len(car_sales))
```

```python
car_sales
```

```python
#Converted Price Column to Numerice
car_sales["Price"] = pd.to_numeric((car_sales["Price"]))

# Get the Cumsum of the Price (Sum of values)
car_sales["Total Sales"] = car_sales["Price"].cumsum();
```

```python
car_sales
```

```python
# Let's plot the total sales

car_sales.plot(x = "Sale Date", y = "Total Sales");
```

```python
#Plot Scatter Plot with Price Coiumn and the Odometer Column
car_sales.plot(x = "Odometer (KM)", y = "Price", kind = "scatter");
```
