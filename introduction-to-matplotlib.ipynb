```python
# 1st method - Using the Object Oriented Method

fig = plt.figure() # creates a figure
ax = fig.add_subplot() # adds some axises

plt.show()
```

```python
# 2nd method - Still using the Object oriented Method....

fig = plt.figure() # Creates a Figure
ax = fig.add_axes([1,1,1,1])

x = [1,2,3,4,5]
y = [11,22,33,44,55]

ax.plot(x,y) # add some data
plt.show()
```

```python
# 3rd Method (Recommended Method...)

fig, ax = plt.subplots()
ax.plot(x, [50,100,200,250,300]) # Add some data...
```

## Matplotlib Example Workflow

```python
# 0. Import Matplotlib and get it ready for plotting....

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# 1. Prepare some data

x = [1,2,3,4]
y = [11, 22, 33, 44]
```

```python
# 2. Setup Plots

fig, ax = plt.subplots(figsize=(10,10))
```

```python
#3. Plot the Data
ax.plot(x,y)
```

```python
#4. Customize the Plot

ax.set(title = "Simple Plot of X & Y Data", xlabel = "X Axis", ylabel = "Y Axis")
```
