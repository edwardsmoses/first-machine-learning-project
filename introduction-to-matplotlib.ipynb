## Which one should you use? (PyPlot vs MatPlotlib OO Method?)

* When plotting something quickly, okay to use the PyPlot method
* When plotting something more advanced, use the OO Method.

```python
## OO Method from Sratch (the former method was mixed with Pyplot style)

fig, ax = plt.subplots(figsize = (10,6))

# Plot the Data
scatter = ax.scatter(x= over_50["age"], y = over_50["chol"], c = over_50["target"]);

# Customize the Plot
ax.set(title = "Heart Disease and Cholesterol Levels", xlabel = "Age", ylabel = "Cholesterol (mg)");

# Add a Legend
ax.legend(*scatter.legend_elements(), title = "Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle = '--');
```

# Tutor's Version - Subplot of Chol, Age, Thalach 

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols =1, figsize = (10,10), sharex=True)

# Add Data to Ax0
scatter = ax0.scatter(x=over_50["age"], y = over_50["chol"], c = over_50["target"]);

#Customize the Axis 0
ax0.set(title = "Heart Disease and Cholesterol Levels", ylabel = "Cholesterol");

#Add a Legend to the Axis 0
ax0.legend(*scatter.legend_elements(), title = "Target");

#Add a Meanline to the Axis 0
ax0.axhline(y=over_50["chol"].mean(), linestyle="--");

# Add a Data to Ax1
scatter = ax1.scatter(x=over_50["age"], y = over_50["thalach"], c = over_50["target"]);

#Customize the Axis 1
ax1.set(title = "Heart Disease and Thalach (Max Heart Rate)", xlabel = "Age", ylabel = "Thalach (Max Heart Rate)");

#Add A Legend to the Axis 1
ax1.legend(*scatter.legend_elements(), title = "Target");

# Add  Meanline to the Axis 1
ax1.axhline(y = over_50["thalach"].mean(), linestyle = "dotted");


#Add a Title to the Figure
fig.suptitle("Heart Disease Analysis", fontsize = 20, fontweight = "bold");
```

# Customization of the X and Y Limitations

```python
plt.style.use("seaborn-talk")

fig, (ax0, ax1) = plt.subplots(nrows=2, ncols =1, figsize = (10,10), sharex=True)

# Add Data to Ax0
scatter = ax0.scatter(x=over_50["age"], y = over_50["chol"], c = over_50["target"], cmap = "summer");

#Customize the Axis 0
ax0.set(title = "Heart Disease and Cholesterol Levels", ylabel = "Cholesterol");

#Change the X Axis Limits
ax0.set_xlim([50,80])

#Add a Legend to the Axis 0
ax0.legend(*scatter.legend_elements(), title = "Target");

#Add a Meanline to the Axis 0
ax0.axhline(y=over_50["chol"].mean(), linestyle="--", color = "red");



# Add a Data to Ax1
scatter = ax1.scatter(x=over_50["age"], y = over_50["thalach"], c = over_50["target"], cmap = "summer");

#Customize the Axis 1
ax1.set(title = "Heart Disease and Thalach (Max Heart Rate)", xlabel = "Age", ylabel = "Thalach (Max Heart Rate)");

#Change ax1 X Axis Limits
ax1.set_xlim([50,80])

#Change the ax1 Y Axis Limits
ax1.set_ylim([60,200])

#Add A Legend to the Axis 1
ax1.legend(*scatter.legend_elements(), title = "Target");

# Add  Meanline to the Axis 1
ax1.axhline(y = over_50["thalach"].mean(), linestyle = "dotted", color = "red");


#Add a Title to the Figure
fig.suptitle("Heart Disease Analysis", fontsize = 20, fontweight = "bold");
```
