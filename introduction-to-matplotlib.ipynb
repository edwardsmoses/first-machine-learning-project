```python
# 2nd method - Still using the Object oriented Method....

fig = plt.figure() # Creates a Figure
ax = fig.add_axes([1,1,1,1])

x = [1,2,3,4,5]
y = [11,22,33,44,55]

ax.plot(x,y) # add some data
plt.show()
```

## Matplotlib Example Workflow

```python
# 0. Import Matplotlib and get it ready for plotting....

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# 1. Prepare some data

x = [1,2,3,4]
y = [11, 22, 33, 44]
```

```python
# 2. Setup Plots

fig, ax = plt.subplots(figsize=(10,10))
```

```python
#3. Plot the Data
ax.plot(x,y)
```

```python
#4. Customize the Plot

ax.set(title = "Simple Plot of X & Y Data", xlabel = "X Axis", ylabel = "Y Axis")
```

```python
fig
```

```python
# 5. Save & Show (Save the Figure to an Image File)
fig.savefig('images/sample-xy-plot.png')
```


## Making figures with NUmpy Arrays

We want:

* Line Plot
* Scatter Plot
* Bar plot
* Historgram
* Subplots


```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
x[:10]
```

```python
# Plot the data and create a line plot... 
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Use the same data to create a scatter plot

fig, ax = plt.subplots()

ax.scatter(x, x**2)
```

```python
fig, ax = plt.subplots()

ax.scatter(x, np.exp(x));
```

```python
# Another scattter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from Dictionary 

nut_butter_prices = {"Almond Butter" : 10, "Peanut Butter": 8, "Cashew Butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())

ax.set(title = "Dan's Nut Butter Store", ylabel = "Price ($)" );
```

```python
# Creating an horizontal Bar 
fig, ax = plt.subplots()

ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data

x = np.random.randn(1000)
```

```python
x[:10]
```

```python
# Create the Histogram using the Data...

fig, ax = plt.subplots()

ax.hist(x)
```


## Two options for creating mutliple sub-plots in the Same Figure...
